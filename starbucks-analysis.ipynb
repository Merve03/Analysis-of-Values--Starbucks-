```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# DataFrame'iniz df olduğunu varsayalım
# Veri tiplerini kontrol edin
print(df.dtypes)

# Sütunları numerik hale getirin, hataları 'NaN' olarak ayarlayın
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Eksik değerleri kontrol edin
print(df.isnull().sum())

# Eksik değerleri doldurun (bu örnekte 0 ile dolduruluyor)
df = df.fillna(0)

# Yalnızca numerik sütunları seçin
df_numeric = df.select_dtypes(include=[np.number])

# Correlation matrix hesaplayın ve ısı haritası oluşturun
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Özellikler ve hedef değişkenler
x = df[["calories", "fat", "carb", "fiber", "protein"]]
y = df["type"]

```

```python
# Veriyi eğitim ve test setlerine ayırma
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Modeli oluşturma ve eğitim
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# Tahminlerde bulunma
y_pred = model.predict(x_test)
```

```python
# Modelin doğruluğunu hesaplama
accuracy = accuracy_score(y_test, y_pred)

print("Modelin doğruluğu:", accuracy)
print("Modelin doğruluğu:", accuracy)
```
